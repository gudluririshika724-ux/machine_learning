```python
#data Integrations visualization and transofrmation 
import pandas as pd 
orders_data=pd.DataFrame({'order_ID':[101,102,103,104],'product_ID':['A','B','C','D'],'Quantity':[2,1,3,2]})

returns_data=pd.DataFrame({'order_ID':[102,104],'Return_reason':['Damaged','wrong product']})

people_data=pd.DataFrame({'Customer_ID':[1,2,3,4],'Customer name':['Alice','Bob','Charli','David']})


merge_data = pd.merge(orders_data, returns_data, on='order_ID', how='left')
final_data= pd.merge(merge_data,people_data,left_on="order_ID",right_on="Customer_ID",how='left')
print(final_data)
```

```python
final_data=pd.DataFrame({
    'order_ID':[101,102,103,104],'product_ID':['A','B','C','D'],'Quantity':[2,1,3,2],
    'Return_reason':['NaN','Damaged','NaN','wrong product'],'Customer_ID':[100,200,300,400],
    'Customer name':['Alice','Bob','Charli','David']
})
final_data.drop(columns=['Customer_ID'],inplace=True)
final_data['Return_reason'].fillna('No return',inplace=True)
final_data['Total_sales']=final_data['Quantity']*10
print(final_data)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create the DataFrame
final_data = pd.DataFrame({
    'order_ID': [101, 102, 103, 104],
    'product_ID': ['A', 'B', 'C', 'D'],
    'Quantity': [2, 1, 3, 2],
    'Return_reason': ['None', 'Damaged', 'None', 'wrong product'],
    'Customer name': ['Alice', 'Bob', 'Charli', 'David']
})

# Step 2: Add Total_sales column (price = 10 per unit)
final_data['Total_sales'] = final_data['Quantity'] * 10

# Step 3: Print to confirm
print(final_data)

# Step 4: Simple bar chart
plt.bar(final_data['product_ID'], final_data['Total_sales'], color='skyblue')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.title('Total Sales per Product')
plt.show()
```

```python
# PCA Visualization of Iris Dataset

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# Step 1: Load the Iris dataset
iris = load_iris()
X = iris.data       # feature data
y = iris.target     # target labels (0,1,2)
target_names = iris.target_names  # ['setosa', 'versicolor', 'virginica']

# Step 2: Apply PCA (reduce to 2 dimensions)
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X)

# Step 3: Create a DataFrame for easy handling
pca_df = pd.DataFrame(data=X_pca, columns=['Principal Component 1', 'Principal Component 2'])
pca_df['target'] = y

# Step 4: Plot the PCA results
plt.figure(figsize=(7, 5))

colors = ['red', 'green', 'blue']
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(
        pca_df.loc[pca_df['target'] == i, 'Principal Component 1'],
        pca_df.loc[pca_df['target'] == i, 'Principal Component 2'],
        color=color,
        label=target_name,
        alpha=0.7
    )

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

tips_dataset =sns.load_dataset("tips")
print(tips_dataset.head())
plt.figure(figsize=(8,6))
sns.scatterplot(x="total_bill", y="tip", data=tips_dataset)


plt.title('Total bill vs Tip Amount')
plt.xlabel('Total bills ($)')
plt.ylabel('Tips ($)')
plt.grid(True)
plt.show()


plt.figure(figsize=(8,6))
sns.histplot(tips_dataset["total_bill"],bins=20, kde=True)
plt.title('Distribution of total bill amount')
plt.xlabel("Total bill")
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

plt.figure(figsize=(8,6))
sns.boxplot(x="day",y="tip",data=tips_dataset)
plt.title("tip amount by day of the week ")
plt.xlabel("Days of the week")
plt.ylabel("Tip Amount ($)")
plt.grid(True)

plt.figure(figsize=(3,3))
sns.barplot(x="day",y="tip",data=tips_dataset)
plt.title("tip amount by day of the week ")
plt.xlabel("Days of the week")
plt.ylabel("Tip Amount ($)")
plt.grid(True)


```
